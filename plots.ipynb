```python
from cleaning_datasets import loading_dissected_datasets
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Plots needed - Moving average plot for each year
def plot_moving_average(year, hour, window=10):
    df = loading_dissected_datasets(year, hour)
    if year < 1986:
        year += 1986
    df = df.reset_index()
    fig = plt.figure()
    plt.plot(df["ITI3"].rolling(window=window).mean())
    plt.xlabel("Days")
    plt.ylabel("Relative energy production")
    plt.title("SMA | {} | {}:00 ".format(year, hour))
```

```python
plot_moving_average(1, 12, window=10)
```

```python
for i in range(0, 24):
    plot_moving_average(1, i, window=10)
```
